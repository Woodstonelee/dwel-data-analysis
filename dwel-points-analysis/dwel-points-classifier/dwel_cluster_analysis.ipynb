# Analyze clusters of DWEL points from KMeans

```python
import itertools

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("darkgrid")
```

```python
cluster_class_files = [ \
                       "/projectnb/echidna/lidar/DWEL_Processing/HF2014/Hardwood20140503/spectral-points-by-union/HFHD20140503-dual-points-clustering/merging/HFHD_20140503_E_dual_cube_bsfix_pxc_update_atp2_ptcl_points_kmeans_clas_0.txt", \
                       "/projectnb/echidna/lidar/DWEL_Processing/HF2014/Hardwood20140503/spectral-points-by-union/HFHD20140503-dual-points-clustering/merging/HFHD_20140503_E_dual_cube_bsfix_pxc_update_atp2_ptcl_points_kmeans_clas_1.txt", \
                       "/projectnb/echidna/lidar/DWEL_Processing/HF2014/Hardwood20140503/spectral-points-by-union/HFHD20140503-dual-points-clustering/merging/HFHD_20140503_E_dual_cube_bsfix_pxc_update_atp2_ptcl_points_kmeans_clas_2.txt"]
cluster_class_markers = [".k", ".r", ".g"]
cluster_class_cmaps = ["Greys", "Reds", "Greens"]
```

```python
data_list = [None, None, None]
for n, fname in enumerate(cluster_class_files):
    data_list[n] = np.genfromtxt(fname, delimiter=',', comments=None, skiprows=3, usecols=(3, 4, 21, 8))
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(16, 16))
for n, marker in enumerate(cluster_class_markers):
    data = data_list[n]
    ax.flat[0].hexbin(data[:, 0], data[:, 1], C=data[:, 3]**2, \
                 reduce_C_function=np.sum, bins='log', alpha=0.33, cmap=plt.get_cmap(cluster_class_cmaps[n]))
    ax.flat[n+1].hexbin(data[:, 0], data[:, 1], C=data[:, 3]**2, \
                 reduce_C_function=np.sum, bins='log', alpha=0.33, cmap=plt.get_cmap(cluster_class_cmaps[n]))
    ax.flat[n+1].set_aspect('equal')
ax.flat[0].set_aspect('equal')
# ax.set_aspect('equal')
```

```python
cluster_class_file = "/projectnb/echidna/lidar/DWEL_Processing/HF2014/Hardwood20140503/spectral-points-by-union/HFHD20140503-dual-points-clustering/merging/HFHD_20140503_C_dual_cube_bsfix_pxc_update_atp2_ptcl_points_kmeans_class_old_-2.txt"
data = np.genfromtxt(cluster_class_file, delimiter=',', comments=None, skiprows=3, usecols=(3, 4, 21, 8))
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(10, 10))
nir_hist = ax1.hist(data[:, 0], bins=1000)
swir_hist = ax2.hist(data[:, 1], bins=1000)
```

```python
print "mean nir = {0:.3f}\n std nir = {1:.3f}\n".format(np.mean(data[:, 0]), np.std(data[:, 0]))
print "mean swir = {0:.3f}\n std swir = {1:.3f}\n".format(np.mean(data[:, 1]), np.std(data[:, 1]))
```
